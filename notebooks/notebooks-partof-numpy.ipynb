```python
import numpy as np
```

```python
points = np.array([
    # [X,   Y ,   Z]
    [1.0, 2.0, 10.0],
    [1.2, 2.1, 10.2],
    [5.0, 6.0, 1.0]
])
print(points)
```

```python
x = points[:, 0]
y = points[:, 1]
z = points[:, 2]

print(x)
```

```python
#Filtering
z_treshold = 5
roof_points = points[z > z_treshold]

print(roof_points)
```
